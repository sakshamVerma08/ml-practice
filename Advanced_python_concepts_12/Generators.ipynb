# Generators
- Generators are special type of functions that don't return the values instantly, instead they return the values when the user asks for the value.
- It produces values one at a time, only when needed, instead of producing entire values together. This helps in Memory Efficiency.

- Normal Function:
```
def get_numbers():
    return [1,2,3] # Returns the entire list together. This list is stored in the memory, and is returned.

get_numbers()
```

- Generator Function:
```
def get_numbers():
    yield 1
    yield 2
    yield 3

iterator = get_numbers()
print(next(iterator)) # prints 1
print(next(iterator)) # prints 2
print(next(iterator)) # prints 3
```

````python
def iterate_ds(iterator):
    try:
        print(next(iterator))
        iterate_ds(iterator)
    
    except StopIteration:
        print("No more elements left to iterate.")

def get_numbers():
    yield 1
    yield 2
    yield 3
    yield 4
    yield 5

iterator = get_numbers()
iterate_ds(iterator)
````

````python
# Use of Generator function inside for loops.
def count():
    yield 1
    yield 2
    yield 3
    yield 4
    yield 5
    yield 6

for i in count():
    print(i)

# Above for loop is internally calling a next() function on the iterator that is created by the generator function. 
# 
````

# Use of 'yield' keyword
- Yield keyword does the following operations:
- Returns the yielded value
- Pauses the function
- next call resumes from where the function was left off (pause point).
