# Closures
- Closures are similar to Nested functions. Basically all closure functions are nested functions, but not all nested functions are closures.
- Nested functions is the superset and closure functions is the subset. It's a concept that uses Nested Functions.
- A Nested function can be described as simply one function containing another function within it.
- Whereas a closure function is a function that uses variables/values from it's enclosing scope, even after the parent function has finished it's execution

```python
# Closures (Nested Functions)
def main_func():

    name = "Saksham Verma"
    print("Hello. This is the main function")

    def sub_func():
        print(name)

    return sub_func()


main_func()

```

```python
def main_func(function):
    name = "Rahul Dravid"

    def sub_function():
        print(name)
        function("name")
        print("This is the sub function.") 

    return sub_function()

main_func(print)


def secondary_func():

    fav_author = "dan brown"

    def print_fav_author():
        print("My fav author = ", fav_author)

    return print_fav_author()
```

# Decorators
- Decorators are used to modify the functionality of a function without touching it's code.
- It's basically used to pass a custom function as an argumen to the declared decorator funtion.
- Decorators work in python because a function in python can be **stored as variable**, **passed as an argument**, **returned from other functions**

```python

@main_func
def custom_func():
    print("I love to learn about new Tech Fields")

@secondary_func
def printAge():
    print("I'm 22 years old")
```

```python
# Very commonly used in Django/Flask/FastAPI etc.
def require_logic(func):

    def wrapper(user):
        if not user["logged_in"]:
            raise Exception("Login Required")
        
        return func(user)
    
    return wrapper


@require_logic
def func(user):
    return "User's profile data"
```

```python
def repeat(n):

    def decorator(func):

        def wrapper(*args,**kwargs):
            for _ in range(n):
                func(*args,**kwargs)
        
        return wrapper 

    return decorator
```

```python
@repeat(7)
def say_hello():
    print("Hello")

say_hello()
```

```python
# Example of Decorator where we are passing the original function as an argument to the decorator function. 

def decorator(function):
    
    def wrapper():
        print("Before function call")
        function()
        print("After function call")

    return wrapper


def greetings():
    print("Hey, Hello, Hi, Hiya, Whasuuup")

greetings = decorator(greetings)
greetings()

# Alternate syntax
''' 
@decorator
def greetings():
    ...

greetings()

''' 
```
