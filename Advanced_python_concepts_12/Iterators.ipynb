# Iterators
- Iterators are advanced python concept that are used to iterate over a data structure in a more Memory efficient way.
- Iterators use Lazy Loading and the use of next() function to iterate over each element of a data structure.

```python
my_list = [1,2,3,4,5,6]

for i in my_list:
    print(i)

print(type(my_list))
```

```python
# Defining an Iterator.
iterator = iter(my_list)
print(iterator)

my_authors = {
    1: "Fyodor dostoevsky",
    2: "Dan Brown",
    3: "J.K Rowling",
    4: "Don Quixote",
    5: "Leo Tolstoy"
}

dict_iterator = iter(my_authors)
print(type(dict_iterator))
```

```python
# Iterating over the Dictionary using try except block.
try:
    print(next(dict_iterator))

except StopIteration:
    print("No more elements left to iterate.")
```

```python
# Iterating over the array using Try-Except Block. 
try:
    print(next(iterator))

except StopIteration:
    print("No more elements left to iterate")
```

```python
# Generic function that works for all Data Structures
def iterate_ds(iterator):
    try:
        print(next(iterator))
        iterate_ds(iterator)

    except StopIteration:
        print("No more elements left to iterate")

name = "Saksham Verma" 
name_iterator = iter(name)
iterate_ds(dict_iterator)
```
